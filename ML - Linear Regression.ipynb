# Linear Regression

```python
# to find the relationship between variables we need to use regression
# In ML, linear regression is used to predict the outcome of future events
```

```python
# draw a scatter plot
import matplotlib.pyplot as plt
x = [10,9,7,4,6,11,22,5,10] # x is age of children
y = [99,85,99,55,57,84,96,82,84] # y is speed of toy cars they are racing

plt.scatter(x, y)
plt.show()
```

```python
# Import scipy and draw the line of linear regression

import matplotlib.pyplot as plt
from scipy import stats

# creating arrays that we want to represent the values of x and y axis
x = [10,9,7,4,6,11,22,5,10] # x is age of children
y = [99,85,99,55,57,84,96,82,84] # y is speed of toy cars they are racing

# execute a method that returns the important key values of linear regression
slope, intercept, r, p, std_err = stats.linregress(x, y)

# creating a function that uses the slope and intercept values to return a new value
# this new value represents where on the y axis the corresponding x value will be placed
def myfunc(x):
    return slope * x + intercept

# running each value for the x array through the function. 
# This will result in a new array with new values for y axis
mymodel = list(map(myfunc, x))

plt.scatter(x, y)
# draw the line of linear regression
plt.plot(x, mymodel)
plt.show()
```

### R for relationship

```python
# r is used for the coefficient of correlation
# its value ranges from -1 to +1
```

### Polynomial Regression

```python
import matplotlib.pyplot as plt
x = [10,9,7,4,6,11,22,5,10] # x is age of children
y = [99,85,99,55,57,84,96,82,84] # y is speed of toy cars they are racing

plt.scatter(x, y)
plt.show()
```

```python
# import numpy and matplotlib then draw a line of the polynomial regression
import numpy
import matplotlib.pyplot as plt

x = [10,9,7,4,6,11,22,5,10] # x is age of children
y = [99,85,99,55,57,84,96,82,84] # y is speed of toy cars they are racing

mymodel = numpy.poly1d(numpy.polyfit(x, y, 3))

myline = numpy.linspace(1, 24, 100)

plt.scatter(x, y)
plt.plot(myline, mymodel(myline))
plt.show()
```

### Multiple Regression

```python
import pandas 
from sklearn import linear_model
df = pandas.read_csv("Tech_Layoffs_of_2022.csv")
```

```python
X = df[['Founded', 'Layoff']]
y = df['Year']

regr = linear_model.LinearRegression()
regr.fit(X, y)

# Predict the Twitter's layoffs given that, it was founded in 2006 where 3700 layoffs have been done during the month of November 2022
predictedCompany = regr.predict([[2006,3700]])

print(predictedCompany)
```
